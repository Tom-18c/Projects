```python
# 滑动提取字符串
A = "123458888867890"
B = "888"

result = []
start = 0
while True:
    idx = A.find(B, start)
    if idx == -1:
        break
    result.append(idx)
    start = idx + 1  # 每次将起始位置向后移动1位，确保重叠的子串也能被找到

print(result)
```
