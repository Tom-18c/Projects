```python
# strip()
# 默认用法 - 去除两端空白
text = "  hello world  "
print(text.strip())  # 输出: "hello world"

# 指定字符集
text = "!!hello world!!"
print(text.strip("!"))  # 输出: "hello world"

# 只去除一端
text = "  hello world  "
print(text.lstrip())  # 只去除左端空白: "hello world  "
print(text.rstrip())  # 只去除右端空白: "  hello world"
```
