```python
# 文件操作
import time
import keyword

user, age = "Tom", 26
# 写入
with open("test.txt", "a", encoding="utf-8") as file_1:
    print(time.strftime("%Y-%m-%d %H:%M:%S"), file=file_1)
    # 格式化输出-%@
    print("我的名字是%s，今年%d岁了" % ("张三", age), file=file_1)
    # 格式化输出-{}
    print("我的名字是{0}，今年{1}岁了".format("李四", age), file=file_1)
    # 格式化输出-f
    print(f"我的名字是{user}，今年{age}岁了", file=file_1)
    print(1, 2, 3, 4, 5, "\n", sep="->", end="\n", file=file_1)
print("数据写入记事本已完成\n")

# 输出-每行显示10个关键字
print("关键字列表：")
max_len = max(len(kw) for kw in keyword.kwlist)
for i in range(0, len(keyword.kwlist), 10):
    print(" | ".join(kw.ljust(max_len) for kw in keyword.kwlist[i : i + 10]))


# 进度条实现，不加 flush=True，可能看不到动态效果，直到循环结束才突然显示 100%
print("\n进度条效果展示：", end="\n")
for i in range(101):
    # \r 让光标回到行首，end="" 阻止换行，flush=True 强制立刻刷新屏幕，显示更流畅
    print(f"\r进度: {i}%", end="", flush=True)
    time.sleep(0.01)
print()  # 循环结束后换行

# \n:换行; \t:制表符; \r:回车; \b:退格; \f:换页; \a:响铃; \e:转义;
```

```python
# 主要内容：数据类型（int、float、complex、bool、str、list、tuple、dict、set、None、type、object）
# 对不可变数据类型，赋值是更改内存地址（id），而不是更改值
# 浮点数截断
float1 = 1.23456
print(f"浮点数截断（四舍五入），a的值是{float1:.3f}")

# 字符串切片，sequence[start:end:step]，左闭右开
str1 = "hello world"
print("字符串切片：", str1[-10:-1:2])

# 列表操作
list_end = 20
list1 = list(range(1, list_end))
print(list1)
list1.append(list_end)
for count_var_1 in range(7):
    list1.append(list_end + 1)
list1.append(50)
print("append，末尾增加内容：", list1)
# 根据内容删除元素-1
for count_var_2 in range(1, 20, 2):
    if count_var_2 in list1:
        list1.remove(count_var_2)
print("remove，删除指定内容（非索引）：", list1)
# 根据内容删除元素-2
for count_var_3 in range(list1.count(list_end + 1)):
    if 11 in list1:
        list1.remove(11)
print(f"remove，删除所有内容为{list_end + 1}的元素：", list1)
```

```python
# 字典
dict1 = {"name": "张三", "age": 18, "sex": "男"}
keys1 = dict1.keys()
values1 = dict1.values()
item1 = dict1.items()
# 返回视图对象
print("字典的键：", keys1)
print(".keys()返回数据类型：", type(keys1))
print("字典的值：", values1)
print(".values()返回数据类型：", type(values1))
print("字典的键值对：", item1)
print(".items()返回数据类型：", type(item1))

for key, value in dict1.items():
    print(f"键: {key}, 值: {value}")
```
