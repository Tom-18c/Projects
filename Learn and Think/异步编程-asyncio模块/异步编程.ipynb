使用python内置模块asyncio实现异步编程

```python
import time
import asyncio


async def task1():
    print("task1 start")
    await asyncio.sleep(4)
    print("task1 end")
    return 10


async def task2(X):
    print("task2 start")
    await asyncio.sleep(3)
    print("task2 end")
    return 20 + X


async def task3():
    print("task3 start")
    await asyncio.sleep(2)
    print("task3 end")
    return 30


async def main():
    # # 创建事件循环
    # event_loop1 = asyncio.get_running_loop()
    # # 手动创建任务，这一步不会进入task函数，只是创建了一个任务对象，通过print结果可以看出
    # result_task1 = event_loop1.create_task(task1())
    # result_task2 = event_loop1.create_task(task2(2))
    # print("等待前")
    # time.sleep(1)
    # print("等待后")
    # result_task3 = event_loop1.create_task(task3())
    # result_task4 = event_loop1.create_task(task2(4))
    # 等待任务完成，并获取结果
    # result1 = await result_task1
    # return await asyncio.gather(result_task1, result_task2, result_task3, result_task4)

    return await asyncio.gather(task1(), task2(2), task3(), task2(4))


if __name__ == "__main__":
    start = time.time()
    # 在常规的python环境中，需要创建事件循环，并运行事件循环
    # loop1 = asyncio.get_event_loop()
    # R = loop1.run_until_complete(main())
    # 等价于
    # R = asyncio.run(main())
    # 在Jupyter Notebook 环境下已经存在运行中的事件循环，所以不需要再创建事件循环
    R = await main()
    print("所有任务最终结果依次为：", R)
    end = time.time()
    print("程序执行时间：", end - start)
```

```python
# 上述精简后
import time
import asyncio


async def task(X):
    print(f"task{X} start")
    await asyncio.sleep(X)
    print(f"task{X} end")
    return 10 + X


async def main():
    return await asyncio.gather(task(4), task(3), task(2), task(1))


if __name__ == "__main__":
    start = time.time()
    R = await main()
    print("所有任务最终结果依次为：", R)
    end = time.time()
    print("程序执行时间：", end - start)
    type_assign = task(5)
    print("赋值操作是将任务对象赋值给变量，并不会立即执行任务函数:", type(type_assign))
    # 在常规的python环境中，需要创建事件循环，并运行事件循环
    # print("使用asyncio.run()方法，可以立即执行任务函数，并返回任务函数的返回值:", asyncio.run(type_assign))
    # 在Jupyter Notebook 环境下已经存在运行中的事件循环
    print("使用await关键字，可以立即执行任务函数，并返回任务函数的返回值:", await type_assign)
```

await 后面跟task、coroutine、future对象，都可以立即执行任务函数，并返回任务函数的返回值

await 后面跟coroutine对象时，仍顺序执行；只有await 后面跟future对象时，才会并发执行
