```python
# 类斐波那契数列生成器
class Fib_Generator:
    def __init__(self, start_list=None):
        # 修复可变默认参数陷阱
        if start_list is None:
            start_list = [0, 1]
        # 将状态绑定到实例属性上
        self.ns = list(start_list)
        self.a = self.ns[-2]
        self.b = self.ns[-1]

    def _format_number(self, num):
        # 消除浮点数精度误差，并处理显示问题
        rounded_num = round(num, 5)
        return int(rounded_num) if rounded_num.is_integer() else rounded_num

    def generate(self, n):
        # 检查输入是否合法，并处理特殊情况
        if not isinstance(n, int) or n < 0:
            raise ValueError("输入错误")
        if n == 0:
            return [self.a, [self.a]]
        elif n == 1:
            return [self.b, [self.a, self.b]]

        # 从第 2 项开始迭代计算，直到第 n 项
        for _ in range(2, n + 1):
            self.a, self.b = self.b, self.a + self.b
            formatted_b = self._format_number(self.b)
            self.ns.append(formatted_b)

        return [self._format_number(self.b), self.ns]
```

以下为实例化的对象调用方法

```python
x = 50
gen1 = Fib_Generator()
print(gen1.generate(x))

gen2 = Fib_Generator([7.2, 7.5])
print(gen2.generate(x))
```
