```python
import numbers


class 提取功能_参数1:
    def __init__(self, 数据源, 特征1):
        self.数据源 = 数据源
        self.特征1 = 特征1

    def 提取特征前数据(self):
        if self.特征1 in self.数据源:
            return self.数据源.split(self.特征1)[0]
        return ""

    def 提取特征后数据(self):
        if self.特征1 in self.数据源:
            return self.数据源.split(self.特征1, 1)[1]
        return ""


class 提取功能_参数2(提取功能_参数1):
    def __init__(self, 数据源, 特征1, 特征2):
        super().__init__(数据源, 特征1)
        self.特征2 = 特征2

    def 提取特征中间数据(self):
        if self.特征1 in self.数据源 and self.特征2 in self.数据源:
            parts = self.数据源.split(self.特征1)
            if len(parts) > 1:
                return parts[1].split(self.特征2)[0]
        return ""


class 格式转化:
    def __init__(self, 数据源):
        self.数据源 = 数据源

    def 文本转数字(self):
        if isinstance(self.数据源, numbers.Number):
            return self.数据源
        if isinstance(self.数据源, str):
            if "." in self.数据源:
                try:
                    return float(self.数据源)
                except ValueError:
                    return False
            else:
                try:
                    return int(self.数据源)
                except ValueError:
                    return False
        return False


a = "9688+8880<520≤2350"

print(提取功能_参数2(a, "8", "≤").提取特征中间数据())
```
