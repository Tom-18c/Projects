```python
# range(11, 20)
class SingleNode:
    def __init__(self, data):
        self.data = data
        self.next = None

    def __iter__(self):
        while self is not None:
            yield self
            self = self.next

    # 可视化
    def __str__(self):
        return "->".join([str(n.data) for n in self])


class DoubleNode(SingleNode):
    def __init__(self, data):
        super().__init__(data)
        self.prev = None

    # 可视化
    def __str__(self):
        return super().__str__().replace("->", "<->")


def linklist(nodes):
    for curr, nxt in zip(nodes, nodes[1:]):
        curr.next = nxt
        if hasattr(nxt, "prev"):
            nxt.prev = curr
```

以下为实例

```python
nodes1 = []
# 创建节点
for var_loop_1 in range(11, 20):
    nodes1.append(SingleNode(var_loop_1))
# 连接节点
linklist(nodes1)


print("单链表：", nodes1[0])


nodes2 = []
# 创建节点
for var_loop_2 in range(21, 30):
    nodes2.append(DoubleNode(var_loop_2))
# 连接节点
linklist(nodes2)


print("双链表：", nodes2[0])
```
