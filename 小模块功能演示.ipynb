```python
# "*"解包功能
my_dict = {"name": "李四"}


def cheak_info(X):
    print("type:\t", type(X))
    print("len:\t", len(X))
    print("value:\t", X)
    print("")


cheak_info(my_dict.items())
cheak_info([*my_dict.items()][0])
cheak_info([*my_dict.items()][0][0])
```

```python
# 列表生成
nums1 = [1, 2, 3, 4]

# 方法1：使用列表推导式
result11 = "-".join([str(n) for n in nums1])
print(result11)  # 输出: 1-2-3-4

# 方法2：使用 map 函数（更简洁优雅）
result12 = "-".join(map(str, nums1))
print(result12)  # 输出: 1-2-3-4
```

```python
# map
# map(函数, 可迭代对象)
def exponentiation_self(x, Ex=2):
    return x**Ex


nums2 = [1, 2, 3, 4]
# 直接调用函数名
result21 = map(exponentiation_self, nums2)
print(list(result21))

# 使用 lambda 表达式
result22 = map(lambda x: exponentiation_self(x, Ex=4), nums2)
print(list(result22))
```

```python
# lambda
# I.当作普通函数调用（匿名函数）
square = lambda x: x**2
print(square(5))  # 输出: 25

# II.结合高阶函数使用（最常见用法）
nums = [1, 2, 3, 4]
# lambda x: x ** 2 会依次接收 nums 中的元素
result = map(lambda x: x**2, nums)
print(list(result))

# III.结合 sorted（排序时指定规则）
people = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35},
]
# lambda p: p["age"] 告诉 sorted 按照字典的 "age" 键来比较大小
sorted_people = sorted(people, key=lambda p: p["age"], reverse=True)
print(sorted_people)

# IV.直接调用（极少用，了解即可）
result = (lambda x, y: x + y)(3, 4)
print(result)

# V.条件表达式
# 虽然 lambda 不能写多行的 if 语句，但支持单行的三元运算符（条件表达式）
# 判断奇偶
check_even = lambda x: "偶数" if x % 2 == 0 else "奇数"
print(check_even(5))  # 输出: 奇数
print(check_even(8))  # 输出: 偶数
```
