```python
from docx import Document
from docx.oxml.ns import qn
import json

docx_path = r"G:\Apps\VSCode\Projects\Base Data\Pyton进阶.docx"
save_json = "word_title_tree.json"


def get_outline_level(para):
    """通过解析 XML 获取段落的大纲级别，如果没有则返回 None"""
    pPr = para.style.element.find(qn("w:pPr"))
    if pPr is not None:
        outlineLvl = pPr.find(qn("w:outlineLvl"))
        if outlineLvl is not None:
            val = outlineLvl.get(qn("w:val"))
            if val is not None:
                return int(val)
    return None


def parse_docx_title_tree(docx_path):
    doc = Document(docx_path)
    root = {"title": "根节点", "children": []}
    # 栈底为根节点，深度为0
    stack = [root]

    for para in doc.paragraphs:
        text = para.text.strip()
        if not text:
            continue

        level = get_outline_level(para)

        # 只有大纲级别存在且在 0~8 之间（对应 Heading 1~9）才视为标题
        if level is not None and 0 <= level <= 8:
            node = {"title": text, "children": []}

            # 栈中保留的元素个数代表当前深度
            # 栈底root深度为0，outlineLvl=0对应Heading 1，深度为1
            # 当遇到深度为 level+1 的标题时，栈中只需保留 level+1 个元素
            while len(stack) > level + 1:
                stack.pop()

            # 将当前节点添加为栈顶节点的子节点
            stack[-1]["children"].append(node)
            # 当前节点入栈，成为新的潜在父节点
            stack.append(node)

        # 正文、缩进、列表等全部跳过
        else:
            continue

    return root


tree = parse_docx_title_tree(docx_path)

with open(save_json, "w", encoding="utf-8") as f:
    json.dump(tree, f, ensure_ascii=False, indent=2)

print("✅ 纯标题结构已保存到：", save_json)
print(json.dumps(tree, ensure_ascii=False, indent=2))
```

```python
from docx import Document
import json

# 配置
docx_path = r"G:\Apps\VSCode\Projects\Base Data\Pyton进阶.docx"
save_json = "word_struct.json"


def parse_docx_tree(docx_path):
    doc = Document(docx_path)
    root = {"title": "根节点", "children": []}
    stack = [root]

    for para in doc.paragraphs:
        text = para.text.strip()
        if not text:
            continue

        style_name = para.style.name.strip()

        # ====================== 关键：适配中文标题样式 ======================
        if style_name == "标题 1":
            node = {"title": text, "children": []}
            while len(stack) > 1:
                stack.pop()
            stack[-1]["children"].append(node)
            stack.append(node)

        elif style_name == "标题 2":
            node = {"title": text, "children": []}
            while len(stack) > 2:
                stack.pop()
            stack[-1]["children"].append(node)
            stack.append(node)

        elif style_name == "标题 3":
            node = {"title": text, "children": []}
            while len(stack) > 3:
                stack.pop()
            stack[-1]["children"].append(node)
            stack.append(node)

        # 正文、缩进正文 → 全部跳过
        else:
            continue

    return root


# 执行解析
tree = parse_docx_tree(docx_path)

# 保存JSON
with open(save_json, "w", encoding="utf-8") as f:
    json.dump(tree, f, ensure_ascii=False, indent=2)

print("✅ 抓取成功！结构已保存：", save_json)
print("\n【抓取结果预览】")
print(json.dumps(tree, ensure_ascii=False, indent=2))
```
